# 3.Check whether all the zones generate the same amount of sales

```python
#oneway Anova
```

H0=Mean sales of Zone-A = Mean sales of Zone-B = Mean sales of Zone-C = Mean sales of Zone-D  

H1=Mean sales of Zone-A != Mean sales of Zone-B != Mean sales of Zone-C != Mean sales of Zone-D

```python
from scipy import stats
stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
```

since p value is less than 0.05 we fail to accept null hypothesis.This means the mean sales of Zone-A != Mean sales of Zone-B != Mean sales of Zone-C != Mean sales of Zone-D .we can conclude that there is a difference in sales among the four zones. In order to tell which groups differed significantly, post-hoc tests need to be conducted.

```python
data.boxplot(column=['Zone - A','Zone - B','Zone - C','Zone - D'],showmeans=True)
plt.show()
```

By default box plots show the median (green line in graph above). The green triangle is the mean for each group.The graphical testing of homogeneity of variances supports the statistical testing findings which is the groups do not have equal variance.
